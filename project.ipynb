```python
# Display the count of matches based on the year wise
import pandas as pd
data = pd.read_csv('Match.csv')
print(data)
```

```python
# Plot a bar graph  no. of matches won by each team in the year  2011
data['Season_Year'].groupby(data.Season_Year).agg('count')
```

```python
def printDataFrameColumns(data):
    li = []
    columns = data.match_winner
    for i in columns:
        li.append(i)
    return li   
printDataFrameColumns(data)
```

```python
x = data[data['Season_Year']==2011]['Season_Year'].groupby(data['match_winner']).count()
x
```

```python
x.index.tolist()
x.values.tolist()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
values = x.values.tolist()
divisions = x.index.tolist()
index = np.arange(len(divisions))

plt.barh(divisions,values,color="purple")
plt.show()
```
